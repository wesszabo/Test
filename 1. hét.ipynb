# 5. feladat és 6.feladat

<div class="exercises">Python nyelven olvass be 5 darab tetszőleges 32x32 pixeles színes képet, jelenítsd meg és írd ki az RGB csatornáit tartalmazó tömbök (mátrixok, tenzorok) tartalmát.</div>

<div class="exercises">Csatornánként számold ki az egyes pixelek átlagát és szórását az 5 képre, majd alakítsd át ezeket $0$ várható értékű, $1$ szórású adathalmazzá.
</div>

## Modulok betöltése

```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.preprocessing import StandardScaler
import scipy.io.wavfile as wavfile
from scipy.fftpack import fft
from bs4 import BeautifulSoup
import requests
import re
import pandas
from collections import Counter
```

## Képek beolvasása és megjelenítése

```python
img = [None] * 5
img[0] = 'Adobe-PDF-Document-icon.png'
img[1] = 'document-excel-icon.png'
img[2] = 'Actions-document-edit-icon.png'
img[3] = 'Files-icon.png'
img[4] = 'folder-documents-icon.png'
title = [None] * 5
title[0] = "First"
title[1] = "Second"
title[2] = "Third"
title[3] = "Fourth"
title[4] = "Fifth"
image = [None] * 5
B = [None] * 5
G = [None] * 5
R = [None] * 5
meanB = [None] * 5
meanG = [None] * 5
meanR = [None] * 5
stdB = [None] * 5
stdG = [None] * 5
stdR = [None] * 5
standardB = [None] * 5
standardG = [None] * 5
standardR = [None] * 5
meanstandardB = [None] * 5
meanstandardG = [None] * 5
meanstandardR = [None] * 5
stdstandardB = [None] * 5
stdstandardG = [None] * 5
stdstandardR = [None] * 5
```

```python
for i in range(0,5):
    plt.subplot(151 + i)
    plt.title(title[i])
    plt.axis('off')
    plt.imshow(mpimg.imread(img[i]))
plt.show()
```

## RGB csatornáit tartalmazó tömbök (mátrixok, tenzorok) tartalma

(RGB csatornát most tanul(tam/om), talán nem értettem félre. Ez is a hiányosságaim közé tartozik, 
aminek a pótlása folyamatban.)

```python
for i in range(0,5):
    image[i] = cv2.imread(img[i])
    B[i], G[i], R[i] = cv2.split(image[i])
```

```python
for i in range(0,5):
    print(title[i])
    print("B csatorna")
    print(B[i])
    print("G csatorna")
    print(G[i])
    print("R csatorna")
    print(R[i])
```

## Csatornánként az egyes pixelek átlaga és szórása

```python
for i in range(0,5):
    meanB[i] = np.mean(B[i], dtype=np.float64)
    meanG[i] = np.mean(G[i], dtype=np.float64)
    meanR[i] = np.mean(R[i], dtype=np.float64)
    stdB[i] = np.std(B[i], dtype=np.float64)
    stdG[i] = np.std(G[i], dtype=np.float64)
    stdR[i] = np.std(R[i], dtype=np.float64)
    print(title[i])
    print("B csatorna átlaga")
    print(meanB[i])
    print("G csatorna átlaga")
    print(meanG[i])
    print("R csatorna átlaga")
    print(meanR[i])
    print("B csatorna szórása")
    print(stdB[i])
    print("G csatorna szórása")
    print(stdG[i])
    print("R csatorna szórása")
    print(stdR[i])
```

## Átalakítás  0 várható értékű,  1  szórású adathalmazzá

```python
scaler = StandardScaler()
for i in range(0,5):
    scaler.fit(B[i])
    standardB[i] = scaler.transform(B[i])
    scaler.fit(G[i])
    standardG[i] = scaler.transform(G[i])
    scaler.fit(R[i])
    standardR[i] = scaler.transform(R[i])  
```

```python
for i in range(0,5):
    meanstandardB[i] = np.mean(standardB[i], dtype=np.float64)
    meanstandardG[i] = np.mean(standardG[i], dtype=np.float64)
    meanstandardR[i] = np.mean(standardR[i], dtype=np.float64)
    stdstandardB[i] = np.std(standardB[i], dtype=np.float64)
    stdstandardG[i] = np.std(standardG[i], dtype=np.float64)
    stdstandardR[i] = np.std(standardR[i], dtype=np.float64)
    print(title[i])
    print("B csatorna átlaga")
    print(meanstandardB[i])
    print("G csatorna átlaga")
    print(meanstandardG[i])
    print("R csatorna átlaga")
    print(meanstandardR[i])
    print("B csatorna szórása")
    print(stdstandardB[i])
    print("G csatorna szórása")
    print(stdstandardG[i])
    print("R csatorna szórása")
    print(stdstandardR[i])
```

# 7. feladat

<div class="exercises">Python nyelven olvass be egy tetszőleges hangfájlt és jelenítsd meg a spektrogramját.
</div>

(Ezt a megoldást az interneten találtam, egy konverziót kellett csak beiktatni. 
Kb. az első ábrát értem. Ez [spektrogram] is a hiányosságaim közé tartozik, idővel ledolgozom.)

```python
myAudio = "bird.wav"

#Read file and get sampling freq [ usually 44100 Hz ]  and sound object
samplingFreq, mySound = wavfile.read(myAudio)

#Check if wave file is 16bit or 32 bit. 24bit is not supported
mySoundDataType = mySound.dtype

#We can convert our sound array to floating point values ranging from -1 to 1 as follows

mySound = mySound / (2.**15)

#Check sample points and sound channel for duel channel(5060, 2) or  (5060, ) for mono channel

mySoundShape = mySound.shape
samplePoints = float(mySound.shape[0])

#Get duration of sound file
signalDuration =  mySound.shape[0] / samplingFreq

#If two channels, then select only one channel
mySoundOneChannel = mySound[:,0]

#Plotting the tone

# We can represent sound by plotting the pressure values against time axis.
#Create an array of sample point in one dimension
timeArray = np.arange(0, samplePoints, 1)

#
timeArray = timeArray / samplingFreq

#Scale to milliSeconds
timeArray = timeArray * 1000

#Plot the tone
plt.plot(timeArray, mySoundOneChannel, color='G')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.show()


#Plot frequency content
#We can get frquency from amplitude and time using FFT , Fast Fourier Transform algorithm

#Get length of mySound object array
mySoundLength = len(mySound)

#Take the Fourier transformation on given sample point 
#fftArray = fft(mySound)
fftArray = fft(mySoundOneChannel)

numUniquePoints = int(np.ceil((mySoundLength + 1) / 2.0))
fftArray = fftArray[0:numUniquePoints]

#FFT contains both magnitude and phase and given in complex numbers in real + imaginary parts (a + ib) format.
#By taking absolute value , we get only real part

fftArray = abs(fftArray)

#Scale the fft array by length of sample points so that magnitude does not depend on
#the length of the signal or on its sampling frequency

fftArray = fftArray / float(mySoundLength)

#FFT has both positive and negative information. Square to get positive only
fftArray = fftArray **2

#Multiply by two (research why?)
#Odd NFFT excludes Nyquist point
if mySoundLength % 2 > 0: #we've got odd number of points in fft
    fftArray[1:len(fftArray)] = fftArray[1:len(fftArray)] * 2

else: #We've got even number of points in fft
    fftArray[1:len(fftArray) -1] = fftArray[1:len(fftArray) -1] * 2  

freqArray = np.arange(0, numUniquePoints, 1.0) * (samplingFreq / mySoundLength);

#Plot the frequency
plt.plot(freqArray/1000, 10 * np.log10 (fftArray), color='B')
plt.xlabel('Frequency (Khz)')
plt.ylabel('Power (dB)')
plt.show()
```

# 8. feladat

<div class="exercises">
Python nyelven töltsd le a <a class="mya" href="http://www.snopes.com/facebook-ai-developed-own-language" target="_blank">Did Facebook Shut Down an AI Experiment Because Chatbots Developed Their Own Language?</a> 
oldal szöveges tartalmát, továbbá hisztogramon jelenítsd meg a tartalomban a betűk előfordulásának gyakoriságát.
</div>

## Oldal letöltése, oldal forráskód megjelenítése

```python
url = "http://www.snopes.com/facebook-ai-developed-own-language"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

## Paragrafusok tartalma 

```python
p_content = soup.find_all('p')
print(p_content)
```

## Paragrafusok száma

```python
number_of_p = len(p_content)
number_of_p
```

## HTML tagek eltávolítása 

```python
for i in range(0, number_of_p):
    p_content[i] = re.sub("<.*?>", "", str(p_content[i]))
    print(p_content[i])
```

## Számok, írásjelek, különleges karakterek eltávolítása

```python
for i in range(0, number_of_p):
    for char in ' @+=?!;:.,"—(‘)’[]”{}“\/&$#%*-0123456789':
        p_content[i] = p_content[i].replace(char, '')
    print(p_content[i])
```

```python
for i in range(0, number_of_p):
    p_content[i] = p_content[i].replace("'", "")
    print(p_content[i])
```

## Nagybetű -> kisbetű konverzió

```python
for i in range(0, number_of_p):
    p_content[i] = p_content[i].lower()
    print(p_content[i])
```

## Összefűzés egyetlen karakterlánccá

```python
content_list = []
for i in range(0, number_of_p):
    content_list.append(p_content[i])
content = ''.join(content_list)
print(content)
```

```python
content
```

```python
content = content.replace("\t", "")
content = content.replace("\n", "")
content = content.replace("\r", "")
content = content.replace("\xa0", "")
```

```python
content
```

## Rendezés abc sorrendbe

```python
content_sorted = ''.join(sorted(content))
content_sorted
```

## Hisztogram

```python
items = content_sorted
letter_counts = Counter(items)
df = pandas.DataFrame.from_dict(letter_counts, orient='index')
df.plot(kind='bar', legend=False, figsize=(15, 20), grid=True, title="Betűk eloszlása")
```
